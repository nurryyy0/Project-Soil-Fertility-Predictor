```python
# ============================================================
# CEK PENGARUH JUMLAH GENERATE KE AKURASI
# ============================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, f1_score
from sklearn.neighbors import NearestNeighbors

RANDOM_STATE = 42
k_neighbors  = 3

df_asli   = pd.read_csv('dataset1.csv')
X_asli    = df_asli.drop('Output', axis=1)
y_asli    = df_asli['Output']
df_kelas2 = df_asli[df_asli['Output'] == 2]

def smote_manual(X, y, target_class, n_samples, k=3, random_state=42):
    np.random.seed(random_state)
    X_min  = X[y == target_class]
    nbrs   = NearestNeighbors(n_neighbors=k+1).fit(X_min)
    _, indices = nbrs.kneighbors(X_min)
    synthetic = []
    for i in range(n_samples):
        idx          = np.random.randint(0, len(X_min))
        neighbor_idx = indices[idx, np.random.randint(1, k+1)]
        diff         = X_min[neighbor_idx] - X_min[idx]
        lam          = np.random.random()
        synthetic.append(X_min[idx] + lam * diff)
    return np.array(synthetic)

# ============================================
# COBA BERBAGAI JUMLAH GENERATE
# ============================================
jumlah_generate = [0, 50,70,80,90,100,110,120,130,150,170,190,200,220,230,240,250,270]
hasil_acc       = []
hasil_f1        = []
hasil_f1_kelas2 = []

for n_gen in jumlah_generate:
    print(f"Testing n_generate = {n_gen}...")

    # Generate data
    if n_gen > 0:
        np.random.seed(RANDOM_STATE)
        df_baru           = df_kelas2.sample(n=n_gen, replace=True,
                                              random_state=RANDOM_STATE).copy()
        noise             = np.random.normal(0, 0.005, df_baru.shape)
        df_baru           = df_baru + noise
        df_baru['Output'] = 2

    # Split dari data asli
    X_train, X_test, y_train, y_test = train_test_split(
        X_asli, y_asli,
        test_size=0.2, random_state=RANDOM_STATE, stratify=y_asli
    )

    # Gabung generate ke train 
    if n_gen > 0:
        df_train           = pd.DataFrame(X_train.values, columns=X_asli.columns)
        df_train['Output'] = y_train.values
        df_train           = pd.concat([df_train, df_baru], ignore_index=True)
        X_train            = df_train.drop('Output', axis=1)
        y_train            = df_train['Output']

    # SMOTE
    X_arr = X_train.values
    y_arr = y_train.values
    mc    = max(Counter(y_arr).values())
    Xr    = X_arr.copy()
    yr    = y_arr.copy()

    for cls in sorted(Counter(y_arr).keys()):
        nn = mc - Counter(y_arr)[cls]
        if nn > 0:
            Xs = smote_manual(X_arr, y_arr, cls, nn)
            Xr = np.vstack([Xr, Xs])
            yr = np.concatenate([yr, [cls]*nn])

    # Train model
    model = Pipeline([
        ("scaler", StandardScaler()),
        ("rf", RandomForestClassifier(
            random_state = RANDOM_STATE,
            class_weight = "balanced",
            n_estimators = 200,
            max_depth    = 10
        ))
    ])
    model.fit(Xr, yr)

    # Evaluasi
    y_pred  = model.predict(X_test.values)
    acc     = accuracy_score(y_test, y_pred)
    f1      = f1_score(y_test, y_pred, average="macro")
    f1_k2   = f1_score(y_test, y_pred, average=None)[2]

    hasil_acc.append(acc)
    hasil_f1.append(f1)
    hasil_f1_kelas2.append(f1_k2)

    print(f"  Test Acc : {acc:.4f} | F1 Macro : {f1:.4f} | F1 Kelas 2 : {f1_k2:.4f}")

# ============================================
# TABEL HASIL
# ============================================
print("\n=====================================================")
print(f"{'Generate':>10} {'Test Acc':>10} {'F1 Macro':>10} {'F1 Kelas2':>10}")
print("-" * 44)
for n, acc, f1, f1k2 in zip(jumlah_generate, hasil_acc, hasil_f1, hasil_f1_kelas2):
    print(f"{n:>10} {acc*100:>9.2f}% {f1:>10.4f} {f1k2:>10.4f}")
print("=====================================================")

best_idx = hasil_f1.index(max(hasil_f1))
print(f"\n✅ Generate terbaik : {jumlah_generate[best_idx]} data")
print(f"   Test Accuracy    : {hasil_acc[best_idx]*100:.2f}%")
print(f"   F1 Macro         : {hasil_f1[best_idx]:.4f}")
print(f"   F1 Kelas 2       : {hasil_f1_kelas2[best_idx]:.4f}")

# ============================================
# VISUALISASI
# ============================================
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].plot(jumlah_generate, [a*100 for a in hasil_acc],
             marker="o", linewidth=2, color="blue")
axes[0].set_title("Test Accuracy")
axes[0].set_xlabel("Jumlah Generate")
axes[0].set_ylabel("Accuracy (%)")
axes[0].grid(True)
for x, y in zip(jumlah_generate, hasil_acc):
    axes[0].annotate(f"{y*100:.1f}%", (x, y*100),
                     textcoords="offset points",
                     xytext=(0, 8), ha="center", fontsize=9)

axes[1].plot(jumlah_generate, hasil_f1,
             marker="o", linewidth=2, color="green")
axes[1].set_title("F1 Macro")
axes[1].set_xlabel("Jumlah Generate")
axes[1].set_ylabel("F1 Score")
axes[1].grid(True)
for x, y in zip(jumlah_generate, hasil_f1):
    axes[1].annotate(f"{y:.3f}", (x, y),
                     textcoords="offset points",
                     xytext=(0, 12), ha="center", fontsize=9)

axes[2].plot(jumlah_generate, hasil_f1_kelas2,
             marker="o", linewidth=2, color="orange")
axes[2].set_title("F1 Kelas 2 (Sangat Subur)")
axes[2].set_xlabel("Jumlah Generate")
axes[2].set_ylabel("F1 Score")
axes[2].grid(True)
for x, y in zip(jumlah_generate, hasil_f1_kelas2):
    axes[2].annotate(f"{y:.3f}", (x, y),
                     textcoords="offset points",
                     xytext=(0, 8), ha="center", fontsize=9)

plt.suptitle("Pengaruh Jumlah Generate ke Akurasi",
             fontsize=13, fontweight="bold")
plt.tight_layout()

plt.show()
```

```python
# RANDOM FOREST + SMOTE ImbPipeline
import pandas as pd
import numpy as np
from scipy.stats import randint
from collections import Counter

from sklearn.model_selection import (
    train_test_split, RandomizedSearchCV,
    StratifiedKFold, cross_val_score
)
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, classification_report,
    confusion_matrix, f1_score
)
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
import joblib
from sklearn.metrics import balanced_accuracy_score

RANDOM_STATE = 42
k_neighbors  = 3


# ─────────────────────────────────────────────
# STEP 1 - LOAD DATA ASLI & DATA GENERATE
# ─────────────────────────────────────────────
df_asli = pd.read_csv('dataset1.csv')
df_baru = pd.read_csv('dataset_gabungan.csv')

print("=== DATASET ===")
print(f"Data asli     : {len(df_asli)} | {dict(df_asli['Output'].value_counts().sort_index())}")
print(f"Data generate : {len(df_baru)} | {dict(df_baru['Output'].value_counts().sort_index())}")


# ─────────────────────────────────────────────
# STEP 2 - SPLIT DARI DATA ASLI
# Test set = data asli murni, tidak pernah kena SMOTE
# ─────────────────────────────────────────────
X_asli = df_asli.drop('Output', axis=1)
y_asli = df_asli['Output']

X_train, X_test, y_train, y_test = train_test_split(
    X_asli, y_asli,
    test_size    = 0.2,
    random_state = RANDOM_STATE,
    stratify     = y_asli         
)

# Gabung data generate HANYA ke train
df_train           = pd.DataFrame(X_train.values, columns=X_asli.columns)
df_train['Output'] = y_train.values
df_train           = pd.concat([df_train, df_baru], ignore_index=True)

X_train = df_train.drop('Output', axis=1)
y_train = df_train['Output']

print(f"\nX_test  (data asli murni) : {Counter(y_test)}")
print(f"X_train (setelah tambah)  : {Counter(y_train)}")


# ─────────────────────────────────────────────
# STEP 3 - HYPERPARAMETER TUNING
# SMOTE di dalam ImbPipeline → hanya kena train fold CV
print("\nMulai Hyperparameter Tuning RF...")

pipeline_rf = ImbPipeline([
    ("smote", SMOTE(
        k_neighbors       = k_neighbors,
        random_state      = RANDOM_STATE,
        sampling_strategy = "not majority" 
    )),
    ("scaler", StandardScaler()),
    ("rf", RandomForestClassifier(
        random_state = RANDOM_STATE,
        oob_score    = True,

    ))
])

param_dist_rf = {
    "rf__n_estimators"     : randint(100, 300),
    "rf__max_depth"        : [3, 5, 7, 10, 12, 15],
    "rf__min_samples_split": randint(10, 30),
    "rf__min_samples_leaf" : randint(5, 15),
    "rf__max_features"     : ["sqrt", "log2"],
    "rf__max_samples"      : [0.6, 0.7, 0.8],
    "rf__criterion"        : ["gini", "entropy"],
    "rf__ccp_alpha"        : [0.001, 0.005, 0.01, 0.02],
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

rand_search = RandomizedSearchCV(
    estimator           = pipeline_rf,
    param_distributions = param_dist_rf,
    n_iter              = 150,
    cv                  = cv,
    scoring             = "f1_macro",
    n_jobs              = -1,
    verbose             = 1,
    random_state        = RANDOM_STATE
)

# Fit dengan data train asli (belum SMOTE)
# SMOTE otomatis jalan hanya di train fold tiap CV iteration
rand_search.fit(X_train.values, y_train.values)

best_model  = rand_search.best_estimator_
best_params = rand_search.best_params_

print("\nBest Params :", best_params)
print("Best CV F1  :", round(rand_search.best_score_, 4))

# Cetak OOB Score dari model terbaik
oob = best_model.named_steps['rf'].oob_score_
print(f"OOB Score   : {oob:.4f}")


# ─────────────────────────────────────────────
# STEP 4 - EVALUASI
# ─────────────────────────────────────────────

# CV di data train (SMOTE hanya kena train fold, val fold tetap bersih)
cv_scores = cross_val_score(
    best_model, X_train.values, y_train.values,
    cv=cv, scoring="f1_macro", n_jobs=-1
)
cv_mean = cv_scores.mean()

# Prediksi di test set (data asli murni, tidak kena SMOTE sama sekali)
y_pred_test = best_model.predict(X_test.values)
test_acc    = accuracy_score(y_test, y_pred_test)
test_f1     = f1_score(y_test, y_pred_test, average="macro")

print("\n--------------------------------")
print(f"CV Train F1 Mean : {cv_mean:.4f}")
print(f"Test F1 Macro    : {test_f1:.4f}")
print(f"Test Accuracy    : {test_acc:.4f}")
print(f"OOB Score        : {oob:.4f}")
train_acc = best_model.score(X_train.values, y_train.values)
print(f"Train Accuracy  : {train_acc:.4f}")
print(f"Best CV Score   : {rand_search.best_score_:.4f}")
bal_acc = balanced_accuracy_score(y_test, y_pred_test)
print(f"Balanced Accuracy : {bal_acc:.4f}")

if (cv_mean - test_f1) < 0.05:
    print("✅ Tidak overfitting")
elif (cv_mean - test_f1) < 0.10:
    print("⚠️  Overfitting ringan")
else:
    print("❌ Overfitting parah")

print("\n--- CLASSIFICATION REPORT ---")
print(classification_report(y_test, y_pred_test))
print("\n--- CONFUSION MATRIX ---")
print(confusion_matrix(y_test, y_pred_test))


# ─────────────────────────────────────────────
# STEP 5 - UJI STABILITAS (10 random seeds)
# ─────────────────────────────────────────────
print("\nUji Stabilitas...")
seeds  = list(range(10))
scores = []

best_rf_params = {
    k.replace("rf__", ""): v
    for k, v in best_params.items()
}

for seed in seeds:
    # Split dari data asli
    Xtr, Xte, ytr, yte = train_test_split(
        X_asli, y_asli,
        test_size=0.2, random_state=seed, stratify=y_asli
    )

    # Gabung generate ke train saja
    df_tr           = pd.DataFrame(Xtr.values, columns=X_asli.columns)
    df_tr['Output'] = ytr.values
    df_tr           = pd.concat([df_tr, df_baru], ignore_index=True)
    Xtr_vals        = df_tr.drop('Output', axis=1).values
    ytr_vals        = df_tr['Output'].values

    # Pipeline dengan SMOTE di dalam (konsisten dengan tuning)
    m = ImbPipeline([
        ("smote", SMOTE(
            k_neighbors       = k_neighbors,
            random_state      = seed,
            sampling_strategy = "not majority"
        )),
        ("scaler", StandardScaler()),
        ("rf", RandomForestClassifier(
            **best_rf_params,
            random_state = seed
         
        ))
    ])
    m.fit(Xtr_vals, ytr_vals)
    scores.append(f1_score(yte, m.predict(Xte.values), average="macro"))

mean_f1 = np.mean(scores)
std_f1  = np.std(scores)
print(f"Seed Scores : {[round(s, 4) for s in scores]}")
print(f"Mean F1     : {mean_f1:.4f}")
print(f"Std F1      : {std_f1:.4f}")


if std_f1 < 0.03:
    print("✅ Model stabil")
elif std_f1 < 0.06:
    print("⚠️  Stabilitas sedang")
else:
    print("❌ Model tidak stabil")


# ─────────────────────────────────────────────
# STEP 6 - SIMPAN MODEL
# ─────────────────────────────────────────────
joblib.dump(best_model, "RF.joblib")
print("\n✅ Model disimpan di RF.joblib")


# ─────────────────────────────────────────────
# STEP 7 - AKURASI KHUSUS PER KELAS
# ─────────────────────────────────────────────
print("\n--- AKURASI PER KELAS (data asli) ---")
for kelas in sorted(y_test.unique()):
    mask      = y_test == kelas
    acc_kelas = accuracy_score(
        y_test[mask],
        best_model.predict(X_test[mask].values)
    )
    print(f"Kelas {kelas} : {acc_kelas:.4f}")
```

```python
# DECISION TREE + SMOTE ImbPipeline

import pandas as pd
import numpy as np
from scipy.stats import randint
from collections import Counter

from sklearn.model_selection import (
    train_test_split, RandomizedSearchCV,
    StratifiedKFold, cross_val_score
)
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, classification_report,
    confusion_matrix, f1_score
)
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
import joblib
from sklearn.metrics import balanced_accuracy_score

RANDOM_STATE = 42
k_neighbors  = 3


# ─────────────────────────────────────────────
# STEP 1 - LOAD DATA ASLI & DATA GENERATE
# ─────────────────────────────────────────────
df_asli = pd.read_csv('dataset1.csv')
df_baru = pd.read_csv('dataset_gabungan.csv')

print("=== DATASET ===")
print(f"Data asli     : {len(df_asli)} | {dict(df_asli['Output'].value_counts().sort_index())}")
print(f"Data generate : {len(df_baru)} | {dict(df_baru['Output'].value_counts().sort_index())}")


# ─────────────────────────────────────────────
# STEP 2 - SPLIT DARI DATA ASLI
# Test set = data asli murni, tidak pernah kena SMOTE
# ─────────────────────────────────────────────
X_asli = df_asli.drop('Output', axis=1)
y_asli = df_asli['Output']

X_train, X_test, y_train, y_test = train_test_split(
    X_asli, y_asli,
    test_size    = 0.2,
    random_state = RANDOM_STATE,
    stratify     = y_asli          
)

# Gabung data generate HANYA ke train
df_train           = pd.DataFrame(X_train.values, columns=X_asli.columns)
df_train['Output'] = y_train.values
df_train           = pd.concat([df_train, df_baru], ignore_index=True)

X_train = df_train.drop('Output', axis=1)
y_train = df_train['Output']

print(f"\nX_test  (data asli murni) : {Counter(y_test)}")
print(f"X_train (setelah tambah)  : {Counter(y_train)}")


# ─────────────────────────────────────────────
# STEP 3 - HYPERPARAMETER TUNING
print("\nMulai Hyperparameter Tuning DT...")

pipeline_dt = ImbPipeline([
    ("smote", SMOTE(
        k_neighbors       = k_neighbors,
        random_state      = RANDOM_STATE,
        sampling_strategy = "not majority"  
    )),
    ("scaler", StandardScaler()),
    ("dt", DecisionTreeClassifier(
        random_state = RANDOM_STATE,
       
    ))
])

param_dist_dt = {
    "dt__max_depth"        : [3, 5, 7, 10, 12, 15, None],
    "dt__min_samples_split": randint(2, 30),
    "dt__min_samples_leaf" : randint(1, 15),
    "dt__max_features"     : ["sqrt", "log2", None],
    "dt__criterion"        : ["gini", "entropy"],
    "dt__splitter"         : ["best", "random"],
    "dt__ccp_alpha"        : [0.0, 0.001, 0.005, 0.01, 0.02],
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

rand_search = RandomizedSearchCV(
    estimator           = pipeline_dt,
    param_distributions = param_dist_dt,
    n_iter              = 150,
    cv                  = cv,
    scoring             = "f1_macro",
    n_jobs              = -1,
    verbose             = 1,
    random_state        = RANDOM_STATE
)

# Fit dengan data train asli (belum SMOTE)
# SMOTE otomatis jalan hanya di train fold tiap CV iteration
rand_search.fit(X_train.values, y_train.values)

best_model  = rand_search.best_estimator_
best_params = rand_search.best_params_

print("\nBest Params :", best_params)
print("Best CV F1  :", round(rand_search.best_score_, 4))


# ─────────────────────────────────────────────
# STEP 4 - EVALUASI
# ─────────────────────────────────────────────

# CV di data train (SMOTE hanya kena train fold, val fold tetap bersih)
cv_scores = cross_val_score(
    best_model, X_train.values, y_train.values,
    cv=cv, scoring="f1_macro", n_jobs=-1
)
cv_mean = cv_scores.mean()

# Prediksi di test set (data asli murni, tidak kena SMOTE sama sekali)
y_pred_test = best_model.predict(X_test.values)
test_acc    = accuracy_score(y_test, y_pred_test)
test_f1     = f1_score(y_test, y_pred_test, average="macro")

print("\n---------------------------------")
print(f"CV Train F1 Mean : {cv_mean:.4f}")
print(f"Test F1 Macro    : {test_f1:.4f}")
print(f"Test Accuracy    : {test_acc:.4f}")
train_acc = best_model.score(X_train.values, y_train.values)
print(f"Train Accuracy  : {train_acc:.4f}")
print(f"Best CV Score   : {rand_search.best_score_:.4f}")
bal_acc = balanced_accuracy_score(y_test, y_pred_test)
print(f"Balanced Accuracy : {bal_acc:.4f}")

print("-----------------------------------")

if (cv_mean - test_f1) < 0.05:
    print("✅ Tidak overfitting")
elif (cv_mean - test_f1) < 0.10:
    print("⚠️  Overfitting ringan")
else:
    print("❌ Overfitting parah")

print("\n--- CLASSIFICATION REPORT ---")
print(classification_report(y_test, y_pred_test))
print("\n--- CONFUSION MATRIX ---")
print(confusion_matrix(y_test, y_pred_test))


# ─────────────────────────────────────────────
# STEP 5 - UJI STABILITAS (10 random seeds)
# ─────────────────────────────────────────────
print("\nUji Stabilitas...")
seeds  = list(range(10))
scores = []

best_dt_params = {
    k.replace("dt__", ""): v
    for k, v in best_params.items()
}

for seed in seeds:
    # Split dari data asli
    Xtr, Xte, ytr, yte = train_test_split(
        X_asli, y_asli,
        test_size=0.2, random_state=seed, stratify=y_asli
    )

    # Gabung generate ke train saja
    df_tr           = pd.DataFrame(Xtr.values, columns=X_asli.columns)
    df_tr['Output'] = ytr.values
    df_tr           = pd.concat([df_tr, df_baru], ignore_index=True)
    Xtr_vals        = df_tr.drop('Output', axis=1).values
    ytr_vals        = df_tr['Output'].values

    # Pipeline dengan SMOTE di dalam (konsisten dengan tuning)
    m = ImbPipeline([
        ("smote", SMOTE(
            k_neighbors       = k_neighbors,
            random_state      = seed,
            sampling_strategy = "not majority"
        )),
        ("scaler", StandardScaler()),
        ("dt", DecisionTreeClassifier(
            **best_dt_params,
            random_state = seed
   
        ))
    ])
    m.fit(Xtr_vals, ytr_vals)
    scores.append(f1_score(yte, m.predict(Xte.values), average="macro"))

mean_f1 = np.mean(scores)
std_f1  = np.std(scores)
print(f"Seed Scores : {[round(s, 4) for s in scores]}")
print(f"Mean F1     : {mean_f1:.4f}")
print(f"Std F1      : {std_f1:.4f}")

if std_f1 < 0.03:
    print("✅ Model stabil")
elif std_f1 < 0.06:
    print("⚠️  Stabilitas sedang")
else:
    print("❌ Model tidak stabil")


# ─────────────────────────────────────────────
# STEP 6 - SIMPAN MODEL
# ─────────────────────────────────────────────
joblib.dump(best_model, "DT.joblib")
print("\n✅ Model disimpan di DT.joblib")


# ─────────────────────────────────────────────
# STEP 7 - AKURASI KHUSUS PER KELAS
# ─────────────────────────────────────────────
print("\n--- AKURASI PER KELAS (data asli) ---")
for kelas in sorted(y_test.unique()):
    mask      = y_test == kelas
    acc_kelas = accuracy_score(
        y_test[mask],
        best_model.predict(X_test[mask].values)
    )
    print(f"Kelas {kelas} : {acc_kelas:.4f}")
```

```python
# SVM + SMOTE 

import pandas as pd
import numpy as np
from scipy.stats import randint, uniform
from collections import Counter

from sklearn.model_selection import (
    train_test_split, RandomizedSearchCV,
    StratifiedKFold, cross_val_score
)
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, classification_report,
    confusion_matrix, f1_score, balanced_accuracy_score
)
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
import joblib

RANDOM_STATE = 42
k_neighbors  = 3


# ─────────────────────────────────────────────
# STEP 1 - LOAD DATA ASLI & DATA GENERATE
# ─────────────────────────────────────────────
df_asli = pd.read_csv('dataset1.csv')
df_baru = pd.read_csv('dataset_gabungan.csv')
df_baru = df_baru[df_baru['Output'] == 2].reset_index(drop=True) 

print("=== DATASET ===")
print(f"Data asli             : {len(df_asli)} | {dict(df_asli['Output'].value_counts().sort_index())}")
print(f"Data generate kelas 2 : {len(df_baru)}")


# ─────────────────────────────────────────────
# STEP 2 - SPLIT DARI DATA ASLI
# Test set = data asli murni, tidak pernah kena SMOTE
# ─────────────────────────────────────────────
X_asli = df_asli.drop('Output', axis=1)
y_asli = df_asli['Output']

X_train, X_test, y_train, y_test = train_test_split(
    X_asli, y_asli,
    test_size    = 0.2,
    random_state = RANDOM_STATE,
    stratify     = y_asli          
)

# Gabung data generate HANYA ke train
df_train           = pd.DataFrame(X_train.values, columns=X_asli.columns)
df_train['Output'] = y_train.values
df_train           = pd.concat([df_train, df_baru], ignore_index=True)

X_train = df_train.drop('Output', axis=1)
y_train = df_train['Output']

print(f"\nX_test  (data asli murni) : {Counter(y_test)}")
print(f"X_train (setelah tambah)  : {Counter(y_train)}")


# ─────────────────────────────────────────────
# STEP 3 - HYPERPARAMETER TUNING
# SMOTE di dalam ImbPipeline → hanya kena train fold CV
# Val fold & test set tidak pernah kena SMOTE
# ─────────────────────────────────────────────
print("\nMulai Hyperparameter Tuning SVM...")

pipeline_svm = ImbPipeline([
    ("smote", SMOTE(
        k_neighbors       = k_neighbors,
        random_state      = RANDOM_STATE,
        sampling_strategy = "not majority" 
    )),
    ("scaler", StandardScaler()),
    ("svm", SVC(
        random_state = RANDOM_STATE,
        probability  = True,
    ))
])

param_dist_svm = {
    "svm__C"       : [0.1, 0.5, 1.0, 2.0, 3.0],
    "svm__kernel"  : ["linear", "rbf"],
    "svm__gamma"   : ["scale", "auto"],
    "svm__tol"     : [1e-3, 1e-4],
    "svm__max_iter": [5000, 10000],
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

rand_search = RandomizedSearchCV(
    estimator           = pipeline_svm,
    param_distributions = param_dist_svm,
    n_iter              = 50,
    cv                  = cv,
    scoring             = "f1_macro",
    n_jobs              = -1,
    verbose             = 1,
    random_state        = RANDOM_STATE
)

# Fit dengan data train asli (belum SMOTE)
# SMOTE otomatis jalan hanya di train fold tiap CV iteration
rand_search.fit(X_train.values, y_train.values)

best_model  = rand_search.best_estimator_
best_params = rand_search.best_params_

print("\nBest Params :", best_params)
print("Best CV F1  :", round(rand_search.best_score_, 4))


# ─────────────────────────────────────────────
# STEP 4 - EVALUASI
# ─────────────────────────────────────────────

# CV di data train (SMOTE hanya kena train fold, val fold tetap bersih)
cv_scores = cross_val_score(
    best_model, X_train.values, y_train.values,
    cv=cv, scoring="f1_macro", n_jobs=-1
)
cv_mean = cv_scores.mean()

# Prediksi di test set (data asli murni, tidak kena SMOTE sama sekali)
y_pred_test = best_model.predict(X_test.values)
test_acc    = accuracy_score(y_test, y_pred_test)
test_f1     = f1_score(y_test, y_pred_test, average="macro")

print("\n--------------------------------")
print(f"CV Train F1 Mean : {cv_mean:.4f}")
print(f"Test F1 Macro    : {test_f1:.4f}")
train_acc = best_model.score(X_train.values, y_train.values)
print(f"Train Accuracy  : {train_acc:.4f}")
print(f"Best CV Score   : {rand_search.best_score_:.4f}")
bal_acc = balanced_accuracy_score(y_test, y_pred_test)
print(f"Balanced Accuracy : {bal_acc:.4f}")

print(f"Test Accuracy    : {test_acc:.4f}")
print("----------------------------------")

if (cv_mean - test_f1) < 0.05:
    print("✅ Tidak overfitting")
elif (cv_mean - test_f1) < 0.10:
    print("⚠️  Overfitting ringan")
else:
    print("❌ Overfitting parah")

print("\n--- CLASSIFICATION REPORT ---")
print(classification_report(y_test, y_pred_test))
print("\n--- CONFUSION MATRIX ---")
print(confusion_matrix(y_test, y_pred_test))


# ─────────────────────────────────────────────
# STEP 5 - UJI STABILITAS (10 random seeds)
# ─────────────────────────────────────────────
print("\nUji Stabilitas...")
seeds  = list(range(10))
scores = []

best_svm_params = {
    k.replace("svm__", ""): v
    for k, v in best_params.items()
}

for seed in seeds:
    # Split dari data asli
    Xtr, Xte, ytr, yte = train_test_split(
        X_asli, y_asli,
        test_size=0.2, random_state=seed, stratify=y_asli
    )

    # Gabung generate ke train saja
    df_tr           = pd.DataFrame(Xtr.values, columns=X_asli.columns)
    df_tr['Output'] = ytr.values
    df_tr           = pd.concat([df_tr, df_baru], ignore_index=True)
    Xtr_vals        = df_tr.drop('Output', axis=1).values
    ytr_vals        = df_tr['Output'].values

    # Pipeline dengan SMOTE di dalam (konsisten dengan tuning)
    m = ImbPipeline([
        ("smote", SMOTE(
            k_neighbors       = k_neighbors,
            random_state      = seed,
            sampling_strategy = "not majority"
        )),
        ("scaler", StandardScaler()),
        ("svm", SVC(
            **best_svm_params,
            random_state = seed,
            probability  = True,
        ))
    ])
    m.fit(Xtr_vals, ytr_vals)
    scores.append(f1_score(yte, m.predict(Xte.values), average="macro"))

mean_f1 = np.mean(scores)
std_f1  = np.std(scores)
print(f"Seed Scores : {[round(s, 4) for s in scores]}")
print(f"Mean F1     : {mean_f1:.4f}")
print(f"Std F1      : {std_f1:.4f}")

if std_f1 < 0.03:
    print("✅ Model stabil")
elif std_f1 < 0.06:
    print("⚠️  Stabilitas sedang")
else:
    print("❌ Model tidak stabil")


# ─────────────────────────────────────────────
# STEP 6 - SIMPAN MODEL
# ─────────────────────────────────────────────
joblib.dump(best_model, "SVM.joblib")
print("\n✅ Model disimpan di SVM.joblib")


# ─────────────────────────────────────────────
# STEP 7 - AKURASI KHUSUS PER KELAS
# ─────────────────────────────────────────────
print("\n--- AKURASI PER KELAS (data asli) ---")
for kelas in sorted(y_test.unique()):
    mask      = y_test == kelas
    acc_kelas = accuracy_score(
        y_test[mask],
        best_model.predict(X_test[mask].values)
    )
    print(f"Kelas {kelas} : {acc_kelas:.4f}")
```
